## Matplotlib workflow example

```python
# 0. Import matplotlib and getting ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #width, height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from a Dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Chasew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's nut butter store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000) 
fig, ax = plt.subplots()
ax.hist(x);
```

### Two option for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2 
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

#Plot to each of different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib Object Oriented (OO) method?)

* When plotting something quickly, it's okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10 ,6))

# Plot the date
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

## Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

## Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```
